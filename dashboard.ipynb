```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
ds=sns.load_dataset("titanic")
```

```python

fig, axes = plt.subplots(4, 3, figsize=(20, 20))

ax = axes.ravel()

# Plot 1: Scatter plot
sns.scatterplot(data=ds, x="age", y="fare", s=30, edgecolor="w", ax=ax[0], hue="sex")
ax[0].set_title("Age vs Fare (Scatter Plot)", fontsize=14, fontweight='bold')
ax[0].set_xlabel("Age", fontsize=12, fontweight='bold')
ax[0].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[0].legend(title='Sex', fontsize=10)

# Plot 2: Line plot
sns.lineplot(data=ds, x="age", y="fare", hue="sex", ax=ax[1])
ax[1].set_title("Age vs Fare (Line Plot)", fontsize=14, fontweight='bold')
ax[1].set_xlabel("Age", fontsize=12, fontweight='bold')
ax[1].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[1].legend(title='Sex', fontsize=10)

# Plot 3: Box plot
sns.boxplot(data=ds, x="class", y="fare", hue="sex", ax=ax[2])
ax[2].set_title("Class vs Fare (Box Plot)", fontsize=14, fontweight='bold')
ax[2].set_xlabel("Class", fontsize=12, fontweight='bold')
ax[2].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[2].legend(title='Sex', fontsize=10)

# Plot 4: Rug plot
survived_counts = ds['survived'].value_counts()
ax[3].pie(survived_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90)
ax[3].set_title("Survival Rate", fontsize=12, fontweight='bold')

# Plot 5: Bar plot
sns.barplot(data=ds, x="class", y="fare", hue="sex", ax=ax[4])
ax[4].set_title("Class vs Fare (Bar Plot)", fontsize=14, fontweight='bold')
ax[4].set_xlabel("Class", fontsize=12, fontweight='bold')
ax[4].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[4].legend(title='Sex', fontsize=10)

# Plot 6: Heatmap
hd = ds.pivot_table(index='class', columns='sex', values='fare', aggfunc='mean')
sns.heatmap(hd, annot=True, cmap='coolwarm', ax=ax[5])
ax[5].set_title("Average Fare by Class and Sex", fontsize=14, fontweight='bold')
ax[5].set_xlabel("Sex", fontsize=12, fontweight='bold')
ax[5].set_ylabel("Class", fontsize=12, fontweight='bold')

# Plot 7: Violin plot
sns.violinplot(data=ds, x="class", y="fare", hue="sex", ax=ax[6], split=True)
ax[6].set_title("Class vs Fare (Violin Plot)", fontsize=14, fontweight='bold')
ax[6].set_xlabel("Class", fontsize=12, fontweight='bold')
ax[6].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[6].legend(title='Sex', fontsize=10)

# Plot 8: Strip plot
sns.stripplot(data=ds, x="class", y="fare", hue="sex", size=5, jitter=True, edgecolor="w", ax=ax[7])
ax[7].set_title("Class vs Fare (Strip Plot)", fontsize=14, fontweight='bold')
ax[7].set_xlabel("Class", fontsize=12, fontweight='bold')
ax[7].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[7].legend(title='Sex', fontsize=10)

# Plot 9: Swarm plot
sns.swarmplot(data=ds, x="class", y="fare", hue="sex", size=5, edgecolor="w", ax=ax[8])
ax[8].set_title("Class vs Fare (Swarm Plot)", fontsize=14, fontweight='bold')
ax[8].set_xlabel("Class", fontsize=12, fontweight='bold')
ax[8].set_ylabel("Fare", fontsize=12, fontweight='bold')
ax[8].legend(title='Sex', fontsize=10)

# Plot 10: Histogram plot
sns.histplot(data=ds, x="fare", kde=False, ax=ax[9])
ax[9].set_title("Fare (Histogram Plot)", fontsize=14, fontweight='bold')
ax[9].set_xlabel("Fare", fontsize=12, fontweight='bold')
ax[9].set_ylabel("Frequency", fontsize=12, fontweight='bold')

# Plot 11: KDE plot
sns.kdeplot(data=ds, x="fare", hue="sex", ax=ax[10])
ax[10].set_title("Fare (KDE Plot)", fontsize=14, fontweight='bold')
ax[10].set_xlabel("Fare", fontsize=12, fontweight='bold')
ax[10].set_ylabel("Density", fontsize=12, fontweight='bold')
ax[10].legend(title='Sex', fontsize=10)

# Plot 12: ECDF plot
sns.ecdfplot(data=ds, x="fare", hue="sex", ax=ax[11])
ax[11].set_title("Fare (ECDF Plot)", fontsize=14, fontweight='bold')
ax[11].set_xlabel("Fare", fontsize=12, fontweight='bold')
ax[11].set_ylabel("ECDF", fontsize=12, fontweight='bold')
ax[11].legend(title='Sex', fontsize=10)



# Adjust layout
plt.tight_layout()
plt.show()
```
