```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titanic = pd.read_csv('titanic.csv')
```

```python
x= titanic['Age']
y= titanic['Fare']
```

```python
plt.scatter(x,y)
```

```python
colors = ['red' if age > 30 else 'blue' for age in x]

plt.scatter(x, y, c=colors)
plt.xlabel('Age (red if > 30)')
plt.ylabel('Fare (green if <= 30)')
plt.title('Scatter Plot of Age vs Fare')
```

```python
plt.scatter(x, y, edgecolor='brown', facecolor='yellow')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

```python
plt.scatter(x, y, c='blue', label='Age')  
plt.scatter(x, y, c='pink', label='Fare', alpha=0.5)  
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.legend(['Age (blue)', 'Fare (pink)'])
```

```python
df = titanic[(titanic['Age'] <= 100) & (titanic['Age'] >= 5) & (titanic['Fare'] <= 400) & (titanic['Fare'] >= 100)]
```

```python
x2 = df['Age']
y2 = df['Fare']
```

```python
colors = ['red' if age > 30 else 'blue' for age in x2]

plt.scatter(x2, y2, c=colors)
plt.xlabel('Age (red if > 30)')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare (Outliers Removed)')
plt.show()
```

```python
x= titanic['Age']
y= titanic['Fare']
colors1 = []
for age in x:
    if age >= 0 and age <= 5:
        colors1.append('yellow')
    elif age >= 6 and age <= 10:
        colors1.append('blue')
    elif age >= 11 and age <= 15:
        colors1.append('green')
    else:
        colors1.append('red')


plt.scatter(x, y, c=colors1)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare with Custom Color Mapping')
plt.show()
```
