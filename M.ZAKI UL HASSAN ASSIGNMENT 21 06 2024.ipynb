```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
# NOW I WILL ASSIGN X TO AGE OF TITANIC DATASET AND Y TO FARE OF TITANIC DATASET
x = df["Age"]
y = df['Fare']
```

```python
# NOW I HAVE CREATED SCATTER PLOT BETWEEN AGE AND FARE
plt.scatter(x,y)
plt.show() #SHOW() IS USED FOR SHOWING DESIRED PLOT
```
