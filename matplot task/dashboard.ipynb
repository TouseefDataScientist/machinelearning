```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

#load dataset
data=pd.read_csv("Iris.csv")

ids=data["Id"]
sepalL=data['SepalLengthCm']
sepalW=data['SepalWidthCm']
petalL=data['PetalLengthCm']
petalW=data['PetalWidthCm']
Species=data['Species']

plt.figure(figsize=(30,18))

plt.subplot(3,5,1)
plt.scatter(sepalL,petalW,color="red")
plt.title("Sepal length vs Petal width")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Petal width (cm)")
plt.grid(True)

#2. Fill b/w : petal l vs petal w
plt.subplot(3,5,2)
plt.fill_between(ids,petalL,petalW,color="green",alpha=0.5)
plt.title("Petal length vs Petal width")
plt.xlabel("Id")
plt.ylabel("length (cm)")
plt.grid(True)

# 3. Stem plot: Sepal l vs sepal w
plt.subplot(3,5,3)
plt.stem(sepalL, sepalW, basefmt=" ")
plt.title("Sepal length vs Sepal width")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")
plt.grid(True)

# 4. stack plot: sepal w vs petal l
plt.subplot(3,5,4)
plt.stackplot(ids,sepalW,petalL, labels=["Sepal width","Petal length"],color=["blue","green"])
plt.legend(loc="upper left")
plt.title("sepal width vs petal length")
plt.xlabel("Id")
plt.ylabel("Value (cm)")
plt.grid(True)

# 5. pie chart: distribution of sepal l
plt.subplot(3, 5, 5)
plt.subplot(3,5,5)
plt.pie(
        sepalL[:10],
        labels=ids[:10],
        shadow=True,
        explode=(0.3,0.2,0,0.10,0,0,0.1,0,0.05,0),
        autopct='%1.1f%%')
plt.title("distribution of sepal length")


# 6: line plot: petal length
plt.subplot(3,5,6)
plt.plot(ids,petalL,color="orange")
plt.title("line plot of petal length")
plt.xlabel("id")
plt.ylabel("petal length(cm)")
plt.grid(True)


# 7: bar plot: sepal l vs petal w
plt.subplot(3,5,7)
plt.bar(sepalL,petalW,color="red")
plt.title("Sepal length vs Petal width")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Petal width (cm)")
plt.grid(True)

# 8: histogram: petal w
plt.subplot(3,5,8)
plt.hist(petalW,bins=15,color="green",label="petal width")
plt.legend()
plt.title("distribution of petal width")
plt.xlabel("petal width(Cm)")
plt.ylabel("frequency")
plt.grid(True)

# 9: box plot: sepal width
plt.subplot(3,5,9)
plt.boxplot(sepalW)
plt.title("box plot of sepal width")
plt.xlabel("index")
plt.ylabel("sepal width(cm)")
plt.grid(True)

#10: violin plot: sepal l by species
plt.subplot(3,5,10)
species=data['Species'].unique()
sepalL_by_species=[data['SepalLengthCm'][data['Species']==sp] for sp in species]
plt.violinplot(sepalL_by_species)
plt.title("Violin plot of sepal length by species")
plt.xlabel("species")
plt.ylabel("sepal length (cm)")
plt.xticks(ticks=[1,2,3], labels=species)

# 11:plot : petal l vs peral w
plt.subplot(3,5,11)
plt.plot(petalL,petalW,color="red")
plt.title("Petal length vs Petal width")
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.grid(True)


#12: area plot:cumulative sepal length
plt.subplot(3,5,12)
plt.fill_between(ids,np.cumsum(sepalL),color="purple",alpha=0.5)
plt.title("Cummulative sepal length")
plt.xlabel("id")
plt.ylabel("cumulative sepal length")
plt.grid(True)

#13: hexabin plot: sepal l vs sepal w
plt.subplot(3,5,13)
plt.hexbin(sepalL,sepalW,gridsize=30,cmap="Blues")
plt.colorbar(label="count")
plt.title("Hexbin plot of sepal length vs sepal width")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width")
plt.grid(True)

# 14:kernel density estimation: petal l
plt.subplot(3,5,14)
sns.kdeplot(petalL, shade=True, color="blue")
plt.title("kde plot of petal lenght")
plt.xlabel("petal length ")
plt.ylabel("density")
plt.grid(True)


#14: swarn plot: sepal w by species
plt.subplot(3,5,15)
sns.swarmplot(x=Species,y=sepalW,data=data)
plt.title("swarn plot of sepal width by species")
plt.xlabel("species")
plt.ylabel("sepal width")
plt.grid(True)

plt.tight_layout()
plt.show()
```
