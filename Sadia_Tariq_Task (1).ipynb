```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Age', color='green')
plt.ylabel('Fare', color='red')
plt.scatter('Age','Fare')
```

```python
x=df['Age']
y=df['Fare']

plt.scatter(x, y, color = 'green')
plt.scatter(y,x, color = 'red')
```

```python
plt.scatter(x,y , c='red', label='Age')
plt.scatter(x,y, c='green', label='Fare', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('scatter plot of Age vs Fare')
plt.legend(['Age(red)','Fare(green)'])
plt.show()
```

```python
plt.scatter(x,y ,edgecolors='red', facecolor='green')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('scatter plot of Age vs Fare')
plt.show()
```

```python
x=df['Age']
y=df['Fare']
colors=['red'if age>30 else 'green'for age in x]
plt.scatter(x,y,c=colors)
plt.xlabel('Age (red if > 30')
plt.ylabel('Fare(green if > 30)')
plt.title('scatter plot of Age vs Fare')
plt.show()
```

```python
colors = ['red' if age <= 12 else 'green' if age <= 19 else 'skyblue' if age <= 30 else 'Orange' for age in x]
plt.scatter(x,y,c=colors)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Relation Between Age and Fare")
plt.legend(['child(red)','Teen(green)','Adult(skyblue)','Others(Orange)'])
plt.show()
```
