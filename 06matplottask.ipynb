```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
x=df['Age']
y=df['Fare']
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c='red',label='Age')
plt.scatter(x,y,c='green',label='Fare',alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(['Age(red)','Fare(green)'])
plt.show()
```

```python
plt.scatter(x,y,edgecolors='red',facecolor='green')

plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
colors = ['red' if age <= 12 else 'green' if age <= 19 else 'blue' if age <= 30 else 'yellow' for age in x]
plt.scatter(x,y,c=colors)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Relation b/w Age and Fare")
plt.legend(['child(red)','Teen(green)','Adult(blue)','Others(yellow)'])
plt.show()
```
