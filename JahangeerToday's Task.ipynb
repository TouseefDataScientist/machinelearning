```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
x=df['Age']
y=df['Fare']
```

```python
# Create a scatter plot with Age as red and Fare as green
plt.scatter(x, y, c='red', label='Age')  # Plot Age in red
plt.scatter(x, y, c='green', label='Fare', alpha=0.5)  # Plot Fare in green with some transparency

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Add legend to explain the colors
plt.legend(['Age (red)', 'Fare (green)'])

plt.show()
```

```python

x = df['Age']
y = df['Fare']


colors = [
    'red' if age < 10 else 'green' if age < 30 else 'blue' if age < 50 else 'yellow' 
    for age in x
]


plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors)
plt.xlabel("age 0- 10 (red), age 10- 30 (green),age 30- 50 (blue),age < 50 (yellow)")
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.legend(['Age 0- 10 (red)', 'Age 10- 30 (green)','Age 30- 50 (blue)','Age < 50 (yellow)'])
plt.show()
```

```python
colors = ['red' if age < 10 else'green' if age < 30 else'blue' if age < 50 else'yellow' for age in x]

legend_labels = ['Age 0-9 (red)', 'Age 10-29 (green)', 'Age 30-49 (blue)', 'Age 50+ (yellow)']
legend_colors = ['red', 'green', 'blue', 'yellow']

legend handle
legend_handles = [plt.Line2D([0, 1], [0, 0], marker='o', color='w', markerfacecolor=color, markersize=10)
                  for color in legend_colors]
plt.legend(legend_handles, legend_labels, loc='best')
plt.scatter(x, y, c=colors)
plt.xlabel('age 0- 10 (red), age 10- 30 (green),age 30- 50 (blue),age < 50 (yellow)')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.figure(figsize=(10, 6))
plt.show()
```
