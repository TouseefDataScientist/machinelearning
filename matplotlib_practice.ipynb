# create label for a plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320])

font1 = {'family': 'arial', 'color': 'red', 'size': 16}
font2 = {'family': 'arial', 'color': 'green', 'size': 12}
font3 = {'family': 'arial', 'color': 'blue', 'size': 26}
font4 = {'family': 'arial', 'color': 'yellow', 'size': 20}
font5 = {'family': 'arial', 'color': 'pink', 'size': 11}
font6 = {'family': 'arial', 'color': 'maroon', 'size': 18}

plt.grid(axis='x', color='red', linestyle= '--', linewidth= 0.5)

plt.plot(x, y)

plt.xlabel("AI")
plt.ylabel("ML", fontdict=font6)
plt.title("Financial Report", fontdict=font1)

plt.show()
```
