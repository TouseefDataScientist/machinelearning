```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('cancer patient data sets.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Level', y='Age', data=data)
plt.title('Age Distribution by Cancer Level')
plt.xlabel('Cancer Level')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Genetic Risk', data=data, color='blue', s=100, alpha=0.8)
plt.title('Age vs. Genetic Risk')
plt.xlabel('Age')
plt.ylabel('Genetic Risk')
plt.grid(True)
plt.show()
```

```python

gender_cancer_counts = data.groupby(['Level', 'Gender']).size().unstack()
plt.figure(figsize=(10, 6))
gender_cancer_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.title('Gender Distribution by Cancer Level')
plt.xlabel('Cancer Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Gender', loc='upper right')
plt.show()
```

```python
cancer_level_counts = data['Level'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(cancer_level_counts, labels=cancer_level_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Cancer Level Distribution')
plt.show()
```

```python
average_age = data.groupby('Level')['Age'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='Level', y='Age', data=average_age, marker='o')
plt.title('Average Age by Cancer Level')
plt.show()

```

```python
symptoms = ['Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring']

# Create a DataFrame to hold the average severity of symptoms by cancer level
symptom_severity = data.groupby('Level')[symptoms].mean().transpose()

plt.figure(figsize=(15, 10))
sns.heatmap(symptom_severity, annot=True, cmap='coolwarm', cbar=True)
plt.title('Average Symptom Severity by Cancer Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='chronic Lung Disease', hue='Level', data=data)  # Use the exact column name
plt.title('Chronic Lung Disease by Cancer Level')
plt.xlabel('Chronic Lung Disease')
plt.ylabel('Count')
plt.legend(title='Cancer Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Level', y='Alcohol use', data=data)
plt.title('Alcohol use Distribution by Cancer Level')
plt.show()
```

```python
selected_columns = ['Age', 'Smoking', 'Alcohol use', 'Balanced Diet', 'Level']
sns.pairplot(data[selected_columns], hue='Level', palette='Set2')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python

# Set up the figure for the dashboard
plt.figure(figsize=(24, 18))
plt.suptitle('Cancer Patient Data Dashboard', fontsize=28)

# Age Distribution Histogram
plt.subplot(3, 3, 1)
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Gender Distribution Bar Plot
plt.subplot(3, 3, 2)
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])

# Age Distribution by Cancer Level Box Plot
plt.subplot(3, 3, 3)
sns.boxplot(x='Level', y='Age', data=data)
plt.title('Age Distribution by Cancer Level')
plt.xlabel('Cancer Level')
plt.ylabel('Age')

# Age vs. Genetic Risk Scatter Plot
plt.subplot(3, 3, 4)
sns.scatterplot(x='Age', y='Genetic Risk', data=data, color='blue', s=100, alpha=0.8)
plt.title('Age vs. Genetic Risk')
plt.xlabel('Age')
plt.ylabel('Genetic Risk')
plt.grid(True)

# Gender Distribution by Cancer Level Stacked Bar Plot
plt.subplot(3, 3, 5)
gender_cancer_counts = data.groupby(['Level', 'Gender']).size().unstack()
gender_cancer_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], ax=plt.gca())
plt.title('Gender Distribution by Cancer Level')
plt.xlabel('Cancer Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Gender', loc='upper right')

# Cancer Level Distribution Pie Chart
plt.subplot(3, 3, 6)
cancer_level_counts = data['Level'].value_counts()
plt.pie(cancer_level_counts, labels=cancer_level_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Cancer Level Distribution')

# Average Age by Cancer Level Line Plot
plt.subplot(3, 3, 7)
average_age = data.groupby('Level')['Age'].mean().reset_index()
sns.lineplot(x='Level', y='Age', data=average_age, marker='o')
plt.title('Average Age by Cancer Level')

# Symptom Severity Heatmap
plt.subplot(3, 3, 8)
symptoms = ['Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring']
symptom_severity = data.groupby('Level')[symptoms].mean().transpose()
sns.heatmap(symptom_severity, annot=True, cmap='coolwarm', cbar=True)
plt.title('Average Symptom Severity by Cancer Level')

# Chronic Lung Disease by Cancer Level Count Plot
plt.subplot(3, 3, 9)
sns.countplot(x='chronic Lung Disease', hue='Level', data=data)
plt.title('Chronic Lung Disease by Cancer Level')
plt.xlabel('Chronic Lung Disease')
plt.ylabel('Count')
plt.legend(title='Cancer Level')

# Display the dashboard
plt.tight_layout()
plt.show()

# Display pair plot separately
selected_columns = ['Age', 'Smoking', 'Alcohol use', 'Balanced Diet', 'Level']
sns.pairplot(data[selected_columns], hue='Level', palette='Set2')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```
