```python
df= pd.read_csv("titanic.csv")
df
```

```python
x=df['Age']
y=df['Fare']


plt.scatter(x,y)
plt.show()
```

```python


plt.scatter(x,y, c='red', label='Age')
plt.scatter(x,y, c='green', label='Fare', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

plt.legend(['Age(red)', 'Fare(green)'])

plt.show() 
```

```python
x=df['Age']
y=df['Fare']


colors=['red' if age <= 10 else'green' if age <=20 else 'blue' if age<= 30 else 'yellow' if age <=40 else 'black' for age in x]

plt.scatter(x,y, c=colors)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

plt.legend(['Age(red)', 'Fare(green)'])
```
